```python
import os
import shutil
from tqdm import tqdm
```

```python
# 1. Rutas originales de tu TFM anterior
ruta_original_imagenes = 'C:/Users/Usuario/Desktop/TFM_NEU/data/train/images'
ruta_original_labels = 'C:/Users/Usuario/Desktop/TFM_NEU/data/train/labels'

# 2. Nueva ruta para nuestro proyecto de Clasificación
ruta_destino = '../data/clasificacion_full'

# Nombres de las clases del NEU-DET
clases = ['crazing', 'inclusion', 'patches', 'pitted_surface', 'rolled-in_scale', 'scratches']

# Creamos las carpetas vacías
for clase in clases:
    os.makedirs(os.path.join(ruta_destino, clase), exist_ok=True)

print("🗂️ Organizando imágenes completas por clase...")

contador = 0
for img_name in tqdm(os.listdir(ruta_original_imagenes)):
    if not img_name.endswith(('.jpg', '.png')): continue
        
    img_path = os.path.join(ruta_original_imagenes, img_name)
    txt_name = img_name.rsplit('.', 1)[0] + '.txt'
    txt_path = os.path.join(ruta_original_labels, txt_name)
    
    # Si no hay txt, saltamos la imagen
    if not os.path.exists(txt_path): continue
        
    # Leemos el txt solo para saber de qué clase es la foto
    with open(txt_path, 'r') as f:
        primera_linea = f.readline().strip()
        if not primera_linea: continue
        
        # Cogemos el primer número (el ID de la clase)
        c_id = int(primera_linea.split()[0])
        clase_nombre = clases[c_id]
        
    # Copiamos LA FOTO ENTERA a la carpeta de su clase
    ruta_guardado = os.path.join(ruta_destino, clase_nombre, img_name)
    shutil.copy(img_path, ruta_guardado)
    contador += 1

print(f"✅ ¡Listo! Se han organizado {contador} imágenes completas en sus carpetas.")
```
